```python
x = (int(input("Enter a number: ")))

if x %2 == 0:
    print(f"{x} is even")
else:
    print(f"{x} is odd")
```

```python
def CheckEvenOdd(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"
    

L = [10, 20,30,45,55]
output= []

for i in L:
    output.append(str(i)+"_"+CheckEvenOdd(i))
print(output)

l = [ i for i in L if CheckEvenOdd(i)=="Odd" ]
print(l)




```
